```python
import pandas as pd
import numpy as np
from google.colab import drive
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Step 1
drive.mount('/content/drive')
file_path = '/content/drive/My Drive/Tugas 2 Python Week 6/data0-SZL.csv'
```

```python
# Step 2
# reading the data
df = pd.read_csv(file_path, header = None )

# checking the dataframe
df.head()
```

```python
# checking data shapes
print(f"ROW : {df.shape[0]}")
print(f"COLUMN : {df.shape[1]}")
```

```python
# Step 3
# computing statistical feature for each column
mean, std, skew, med, en, min, max = [], [], [], [], [], [], []
for i in range(df.shape[1]):

    compute_mean = df[i].mean()
    mean.append(compute_mean)

    compute_std = df[i].std()
    std.append(compute_std)

    compute_skew = df[i].skew()
    skew.append(compute_skew)

    compute_med = df[i].median()
    med.append(compute_med)

    compute_energy = (df[i]**2).sum()
    en.append(compute_energy)

    compute_min = df[i].min()
    min.append(compute_min)

    compute_max = df[i].max()
    max.append(compute_max)

# input statistics to dataframe
stat_features = ['mean', 'std', 'skew',
                 'med', 'energy', 'min', 'max']

df_stat = pd.DataFrame()

df_stat['mean'] = mean
df_stat['std'] = std
df_stat['skew'] = skew
df_stat['med'] = med
df_stat['energy'] = en
df_stat['min'] = min
df_stat['max'] = max

df_stat.head()
```

```python
# Step 4
fig, ax = plt.subplots(2, 3, figsize = (18,8), dpi = 300)


ax[0][0].plot(df.index, df[100])
ax[0][0].set_title("Kolom 100")

ax[0][1].plot(df.index, df[101])
ax[0][1].set_title("Kolom 101")

ax[0][2].plot(df.index, df[18])
ax[0][2].set_title("Kolom 18")

ax[1][0].plot(df.index, df[88])
ax[1][0].set_title("Kolom 88")

ax[1][1].plot(df.index, df[54])
ax[1][1].set_title("Kolom 54")

ax[1][2].plot(df.index, df[32])
ax[1][2].set_title("Kolom 32")



plt.show()
```

```python
# Step 5
#setting the color and marker for plotting

color = ['red', 'blue', 'green', 'orange', 'purple',
         'brown', 'pink', 'gray', 'cyan','crimson',
         'black', 'lightcoral','navy', 'yellow', 'indigo',
         'deeppink']
marker = ['1', '2', '3','4', '+',
          'x', '|', '_', 0, 1,
          2, 3, 4, 5, 6, 7]

plt.figure(figsize=(18,10), dpi = 300)
for i in range(0, df.shape[1], 10):
    x = df_stat[i:i+10]['mean']
    y = df_stat[i:i+10]['std']

    plt.scatter(x,y, c = color[i // 10], marker = marker[i // 10],
                label = f"Kolom {i} - {i+10}")

plt.legend()
plt.xlabel("Rata-rata")
plt.ylabel('Standard Deviasi')
plt.title("Rata-rata vs. Standard Deviasi")
plt.show()
```

```python
# Step 6
plt.figure(dpi = 300, figsize = (18, 10))
for i in range(0, df.shape[1], 10):
    x = df_stat[i:i+10]['skew']
    y = df_stat[i:i+10]['med']
    plt.scatter(x,y, c = color[i // 10], marker = marker[i // 10],
                label = f'Kolom {i} - {i+10}')
plt.legend()
plt.xlabel("Skewness")
plt.ylabel("Median")
plt.title("Skewness vs. Median")
plt.show()
```
